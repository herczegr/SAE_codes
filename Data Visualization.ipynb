```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# n = 500
data = pd.read_excel('~/Documents/Survey statisztika és adatanalitika 2021:23/Szakdoga/adatvizu_val.xlsx')
data2 = pd.read_excel('~/Documents/Survey statisztika és adatanalitika 2021:23/Szakdoga/viz_val2.xlsx')
```

```python
boxprops = dict(facecolor='lightblue', color='darkred')
medianprops = dict(color='darkred')
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(11, 10))


bplot1 = axes[0].boxplot(data,
                         vert=True,   
                         patch_artist=True,
                         showfliers=False,
                         boxprops=boxprops,   
                         medianprops=medianprops)   

bplot2 = axes[1].boxplot(data2,
                         vert=True,  
                         patch_artist=True,
                         showfliers=False,
                         boxprops=boxprops,   
                         medianprops=medianprops)   

axes[0].set_xlabel('Domain')
axes[0].set_ylabel('Gross Personal Income')
axes[1].set_xlabel('Domain')
axes[1].set_ylabel('Gross Personal Income')
axes[0].set_ylim(-2000, 14000)
axes[1].set_ylim(-2000, 14000)
```

HEATMAP (Chapter 5.1.1)

```python
hmap = pd.read_excel('~/Documents/Survey statisztika és adatanalitika 2021:23/Szakdoga/populacio_d.xlsx')
hmap.head()
```

```python
pivot_table = pd.pivot_table(data=hmap, index='age_group', columns='highest_edu', aggfunc='size')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='coolwarm', cbar_kws={'label': 'Count'}, linewidths=0.5)

plt.xlabel('Educational Level')
plt.ylabel('Age Group')
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```
